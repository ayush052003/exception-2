```python
'''Q3. What errors are defined in the Arithmetic Error class? Explain any two with an example.'''
```

```python
# The ArithmeticError Exception is the base class for all errors associated with arithmetic operation.

# Some errors associated with the arithmetic error are...
# 1.ZeroDivisionError
# 2.OverflowError
# 3.FloatingPointError
```

```python
# ZeroDivisionError

try:
    20/0
except ZeroDivisionError as e:
    print(e)
```

```python
# OverflowError

j = 5.0
try:
    for i in range(1, 1000):
        j = j**i
except OverflowError as e:
    print(e)
```

```python
'''lookupError Exception is the Base class for errors raised when something can’t be found. The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: IndexError, KeyError.'''
```

```python
# IndexError - An IndexError is raised when a sequence reference is out of range

try:
    l=[1,2,3,4,5]
    print(l[6])
except IndexError as e:
    print(e)
```

```python
# KeyError - A KeyError is raised when a value is not found as a key of a dictionary. The given code is rewritten as follows to catch the exception and find its type.

try:
    
    d={"key":"ayush",1:[2,3,4,5]}
    print(d["key2"])
except KeyError as e:
    print(e)
```
